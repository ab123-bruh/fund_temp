```python
import yfinance as yf
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython import get_ipython
from scipy.stats import spearmanr
```

```python
gld = yf.download('GLD', period='5y', interval='1mo')
gld = gld.rename(columns={'Close' : 'GLD'})
tlt = yf.download('TLT', period='5y', interval='1mo')
tlt = tlt.rename(columns={'Close' : 'TLT'})
inda = yf.download('INDA', period='5y', interval='1mo')
inda = inda.rename(columns={'Close' : 'INDA'})
ewq= yf.download('EWQ', period = '5y', interval='1mo')
ewq = ewq.rename(columns={'Close': 'EWQ'})





corr_df = pd.concat([gld.GLD, tlt.TLT, inda.INDA, ewq.EWQ], axis = 1)
corr_matrix= corr_df.corr(method='spearman')

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix - Macro')
plt.show()
```
